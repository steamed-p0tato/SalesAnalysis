```python
#### now concatenate the 12 months data in a single data frame
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir("./Sales_Data")]

allmonthsdata = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    
    allmonthsdata = pd.concat([allmonthsdata, df])


allmonthsdata.to_csv("alldata.csv", index =False)
```

```python
dx = pd.read_csv("alldata.csv")
dx.head()
```

## clean up the data!

```python
nan_df = dx[dx.isna().any(axis=1)]
nan_df.head()
```

```python
dx = dx.dropna(how="all")
```

### finding the "or" and removing it

```python
or_df = dx[dx["Order Date"].str[0:2] == "Or"]
or_df.head()
```

```python
dx = dx[dx["Order Date"].str[0:2] != "Or"]   #!= removes all rows with "Or"
```

### adding columns

```python
dx["month"] = dx["Order Date"].str[0:2]
dx["month"] = dx["month"].astype('int32')
dx.head()
```

## best month for sale and how much earned??

#### adding a sales column to multiply qty and price

```python
# changing the data type of the columns

dx['Quantity Ordered'] = pd.to_numeric(dx['Quantity Ordered'])
dx["Price Each"] = pd.to_numeric(dx["Price Each"])
```

```python
dx["sales"] = dx["Quantity Ordered"] * dx["Price Each"]
dx.head()
```

```python
results = dx.groupby("month").sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)  #for 12 months

plt.bar(months, results["sales"])

plt.show
```

### what city has highest no. of sales

#### add a city column

```python
dx["state"] = 1
dx.head()
```

```python
def getcity(address):
    return address.split(',')[1]

def getstate(address):
    return address.split(',')[2].split(' ')[1]
```

```python
# use .apply func

dx["state"] = dx["Purchase Address"].apply(lambda x: getcity(x) + ' ' +getstate(x))
dx.head()
```

### grouping

```python
results = dx.groupby("state").sum()
```

```python
results
```

```python
import matplotlib.pyplot as plt

state = dx['state'].unique()
plt.bar(state, results["sales"])
plt.xticks(state, rotation='vertical', size = 8)



plt.show()
```

###  !!! as we can see san franc must have been the top but isnt so we have to put dx in the same  X, Y as its group by for that do a for loop for state in the dataframe of dx-groupby

```python
import matplotlib.pyplot as plt

state = [state for state, df in dx.groupby('state')]
plt.bar(state, results["sales"])
plt.xticks(state, rotation='vertical', size = 8)



plt.show()
```
